```python
df = pd.read_csv('../../Datasets/car price/CarPrice_Assignment.csv')
df = df.drop('car_ID', axis=1)
df = df.drop('CarName', axis=1)
```

```python
df.head()
```

```python
binary_columns = []
categorical_columns = []

for c in df.columns:
    if df[c].dtype == 'object':
        if len(df[c].value_counts()) == 2:
            binary_columns.append(c)
        if len(df[c].value_counts()) > 2:
            categorical_columns.append(c)
```

```python
print(binary_columns)
print(categorical_columns)
```

```python
df = pd.get_dummies(df, prefix=binary_columns, columns=binary_columns , drop_first=True)
df = pd.get_dummies(df, prefix=categorical_columns, columns=categorical_columns)
```

```python
df.head()
```

```python
y = df.price.to_numpy()
df = df.drop('price', axis=1)
X = df.to_numpy()
```

```python
p = np.random.permutation(X.shape[0])

X = X[p]
y = y[p]
```

```python
split1 = int(X.shape[0] * 0.6)
split2 = int(X.shape[0] * 0.8)

X_train = X[:split1]
X_val = X[split1:split2]
X_test = X[split2:]

y_train = y[:split1]
y_val = y[split1:split2]
y_test = y[split2:]
```

```python
print(X_train.shape, X_val.shape, X_test.shape)
```

```python
mean_train = X_train.mean(axis=0)
std_train = X_train.std(axis=0)

X_train = (X_train - mean_train) / std_train
X_val = (X_val - mean_train) / std_train
X_test = (X_test - mean_train) / std_train
```

```python
np.save('X_train.npy', X_train)
np.save('X_val.npy', X_val)
np.save('X_test.npy', X_test)
```
